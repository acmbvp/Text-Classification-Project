# Creating training and testing dataframes

```python
import numpy as np
import pandas as pd
import math
from sklearn.naive_bayes import MultinomialNB
import string
import os
```

```python
path = "E:\\ML\\Text Classification Data\\data\\20_newsgroups_train"
```

```python
stop_words = open("E:\\ML\\Text Classification Data\\stopwords.txt","r").read().split()
len(stop_words)
```

```python
total_words = []
counts = {}

for filefolder in os.listdir(path):
    for filename in os.listdir(path + "\\" + filefolder):
        
        f = open((path + "\\" + filefolder + "\\" + filename),"r")
        file = f.read()
        
        file = file[file.find('Lines')+10:]
        file = file.lower()
        
        for c in (string.punctuation + string.digits):
            
            if c == "'":
                continue
            
            file = file.replace(c," ")
            
        file_words = file.split()
        
        words = []
        for word in file_words:
            
            if len(word) > 3:
                if word.isdigit() == False:
                    if word not in stop_words:
                        words.append(word)
                        if word in counts:
                            counts[word] += 1
                        else:
                            counts[word] = 1
        
        
        total_words.extend(words)
```

```python
print("Total words =", len(total_words))
print("Set of Total words =", len(set(total_words)))
```

## Creating vocabulory

```python
sorted_by_value = sorted(counts.items(), key = lambda kv: kv[1], reverse=True)

sorted_sliced = sorted_by_value[0:6000]         # 6000 words

vocabulory = []
for item in sorted_sliced:
    vocabulory.append(item[0])

print(len(vocabulory))
vocabulory
```

## Creating training dataframe

```python
data = []
output = []

for filefolder in os.listdir(path):
    for filename in os.listdir(path + "\\" + filefolder):
        
        row_data = []
        
        f = open((path + "\\" + filefolder + "\\" + filename),"r")
        file = f.read()
        
        file = file[file.find('Lines')+10:]
        file = file.lower()
        
        for c in (string.punctuation + string.digits):
            
            if c == "'":
                continue
            
            file = file.replace(c," ")
            
        file_words = file.split()
        
        words = []
        for word in file_words:
            
            if len(word) > 3:
                if word.isdigit() == False:
                    if word not in stop_words:
                        words.append(word)
        
        for word in vocabulory:
            
            count = words.count(word)
            row_data.append(count)
            
        data.append(row_data)
        
        output.append(filefolder)
```

```python
dfa = np.array(data).reshape(len(data), len(vocabulory))
df = pd.DataFrame(dfa)
df.columns = vocabulory
df["Output"] = output
df
```

```python
df.to_csv("TrainDataFrame.csv")                # Creating a csv file of the dataframe
```

```python
test_path = "E:\\ML\\Text Classification Data\\data\\20_newsgroups_test"
```

## Creating testing dataframe

```python
test_data = []
test_output = []

for filefolder in os.listdir(test_path):
    for filename in os.listdir(test_path + "\\" + filefolder):
        
        row_data = []
        
        f = open((test_path + "\\" + filefolder + "\\" + filename),"r")
        file = f.read()
        
        file = file[file.find('Lines')+10:]
        file = file.lower()
        
        for c in (string.punctuation + string.digits):
            
            if c == "'":
                continue
            
            file = file.replace(c," ")
            
        file_words = file.split()
        
        words = []
        for word in file_words:
            
            if len(word) > 3:
                if word.isdigit() == False:
                    if word not in stop_words:
                        words.append(word)
        
        for word in vocabulory:
            
            count = words.count(word)
            row_data.append(count)
            
        test_data.append(row_data)
        
        test_output.append(filefolder)
```

```python
test_dfa = np.array(test_data).reshape(len(test_data), len(vocabulory))
test_df = pd.DataFrame(test_dfa)
test_df.columns = vocabulory
test_df["Output"] = test_output
test_df
```

```python
test_df.to_csv("TestDataFrame.csv")                 # Creating a csv file of the dataframe
```

```python
x_train = df[df.columns[0:-1]]
y_train = df[df.columns[-1]]
x_test = test_df[test_df.columns[0:-1]]
y_test = test_df[test_df.columns[-1]]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```
